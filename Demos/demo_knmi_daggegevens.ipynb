```python
from urllib.parse import urlparse

import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
year = 2024

start = f'{year}-01-01'
end = f'{year}-12-31'

stations = {
    260: 'De Bilt',
    279: 'Hoogeveen',
}

stns = ':'.join(map(str, stations.keys()))
vars = ':'.join(['TG','TN','TX'])

url = 'https://www.daggegevens.knmi.nl/klimatologie/daggegevens'

url += f'?start={start}'
url += f'&end={end}'
url += f'&stns={stns}'
url += f'&vars={vars}'

print(url)
```

```python
n_header_lines = 6 + len(stns.split(':')) + len(vars.split(':'))

df = pd.read_csv(url,
                 skiprows=n_header_lines, 
                 skipinitialspace=True)

df.head()
```

```python
df.rename(columns={'# STN': 'STN'}, inplace=True)

df[['TG', 'TN', 'TX']] /= 10

df['dd'] = pd.to_datetime(df['YYYYMMDD'], format='%Y%m%d')
df = df.set_index('dd')

df
```
